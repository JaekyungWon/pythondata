```python
import folium
import pandas as pd
import json
import requests
```

```python
df = pd.read_csv('sasangWater.csv', encoding='cp949')
df
```

```python
def get_geo_code(query):
    RESTAPI = '706b841e9c08dc6e0427fde5fd745062'
    url = f"https://dapi.kakao.com/v2/local/search/address.json?query={query}"
    headers ={"Authorization":f"KakaoAK {RESTAPI}"}
    res = requests.get(url,headers=headers)
    res = res.json()
    
    try:
        return [float(res['documents'][0]['y']),float(res['documents'][0]['x'])]
    except:
        return "결과 없음"
```

```python
get_geo_code('부산광역시 사상구 광장로 93번길 6')
```

```python
df['소재지 주소'] = df['소재지 주소'].str.split('-').str.get(0)
```

```python
df['소재지 주소']
```

```python
df['addr'] = df['소재지 주소'].apply(get_geo_code)
df
```

```python
from folium.plugins import MarkerCluster

map = folium.Map([35.1542775, 128.9794013])
marker_cluster = MarkerCluster().add_to(map)

for index, row in df.iterrows():
    if row['addr'] != '결과 없음':
        folium.Marker(row['addr'], popup = row['민방위 비상급수시설명'], tooltip=row['소재지 주소']).add_to(marker_cluster)
        
map
```
