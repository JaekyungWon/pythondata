```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
```

```python
# 마이너스기호(-)가 깨질때 사용하면 됨
mpl.rcParams['axes.unicode_minus'] = False
```

```python
data = np.random.randint(-100,100,50).cumsum()
plt.plot(range(50),data,'r')
plt.title('시간별 가격 추이')
plt.ylabel('주식가격')
plt.xlabel('시간(분)')
```
